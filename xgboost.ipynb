```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import PredefinedSplit
import numpy as np
from xgboost import XGBRegressor
import wandb
```

```python
from data_utils.feature_utils import create_features_from_raw_df
from data_utils.scoring_utils import score_spearman
```

```python
raw_df = pd.read_csv("data/GDPa1_v1.2_20250814.csv")
feature_df = create_features_from_raw_df(raw_df).drop("vh_turn", axis=1)
y = raw_df['HIC']
splitter = PredefinedSplit(raw_df['hierarchical_cluster_IgG_isotype_stratified_fold'])
```

```python
def fit_to_all_folds(X, y, splitter: PredefinedSplit, model, scorer):
'''
    Fits a given model to train and test data across folds specified by an sklearn splitter.
'''
    scores = []
    for i, (train_idx, test_idx) in enumerate(splitter.split()):
        X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
        y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]

        # Filter na
        model.fit(X_train.loc[~y_train.isna()], y_train.loc[~y_train.isna()])
        scores.append(scorer(model, X_test.loc[~y_test.isna()], y_test.loc[~y_test.isna()]))

    mean_scores = []
    for score in scores:
        if np.isnan(score):
            mean_scores.append(0)
        else:
            mean_scores.append(score)
    return np.mean(mean_scores), scores
```

```python
model = RandomForestRegressor(
        n_estimators=1000,
        min_samples_leaf=1,
        #max_features='log2',
        criterion='friedman_mse',
        ccp_alpha=0
    )
```

```python
def main():
    with wandb.init(project="Antibody Succotash") as run:
        model.set_params(**run.config)
        score, fold_scores = fit_to_all_folds(feature_df, y, splitter, model, score_spearman)
        logs = {"model": "Random_Forest", "spearman": score}
        logs.update({f"fold-{i}-score": s for i, s in enumerate(fold_scores)})
        run.log(logs)

# 2: Define the search space
sweep_configuration = {
    "method": "bayes",
    "metric": {"goal": "maximize", "name": "spearman"},
    "parameters": {
        "min_samples_leaf": {"max": 1-1e-3, "min": 1e-3},
        "max_features": {"values":[None, "log2", "sqrt"]},
        "max_depth": {"max": 10, "min":1, "distribution": "int_uniform" }
    },
}

# 3: Start the sweep
sweep_id = wandb.sweep(sweep=sweep_configuration, project="Antibody Succotash")

wandb.agent(sweep_id, function=main, count=100)
```
